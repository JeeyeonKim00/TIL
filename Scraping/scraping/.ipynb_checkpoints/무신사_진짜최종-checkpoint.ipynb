```python
def get_chrome_driver():
    # 1. 브라우저 옵션 세팅
    chrome_options = webdriver.ChromeOptions()
    
    # 2. driver 생성
    driver = webdriver.Chrome(
            service = Service(ChromeDriverManager().install()),
            options = chrome_options)
    return driver
```

# 지금까지 추출과정 한번에 묶기

```python
from selenium.webdriver.common.by import By

MUSINSA_URL ="https://www.musinsa.com/categories/item/021"

driver = get_chrome_driver()
driver.get(MUSINSA_URL)


goods_list= driver.find_elements(By.CSS_SELECTOR, '#searchList div.li_inner ')

all_data=[]

for i in range(len(goods_list)):
    product_code  = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.txt_cnt_like').get_attribute('data-goodsno')
    brand = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.item_title > a').text
    product_name = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.list_info > a').text
    price = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.price').text.split(' ')[-1]
    try:
        star = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.point > span:nth-child(2)').text
    except:
        star= None
        
    try:
        like = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.txt_cnt_like ').text
    except:
        like = None
    
    detail_url = goods_list[i].find_element(By.CSS_SELECTOR, 'div.article_info > p.list_info >a ').get_attribute('href')
    all_data.append([product_code,brand,product_name,price,star,like,detail_url])
```

# 위의 과정 하나로 만들기

```python
driver = get_chrome_driver()
season_list = []
gender_list=[]
detail_img_list=[]
for i in range(len(all_data)):
    url = all_data[i][5]
    driver.get(url)
    try:
        season = driver.find_element(By.CSS_SELECTOR,'#product_order_info ul.product_article>li:nth-child(2)>p.product_article_contents>strong').text
    except:
        season =None
    gender = driver.find_element(By.CSS_SELECTOR,'#product_order_info ul.product_article>li:nth-child(2)>p.product_article_contents>span.txt_gender').text
    img = driver.find_elements(By.CSS_SELECTOR,'#detail_thumb > ul >li')
    for j in range(len(img)):
        detail_img_url=img[j].find_element(By.CSS_SELECTOR, 'img ').get_attribute('src')
    
    season_list.append(season)
    gender_list.append(gender)
    detail_img_list.append(detail_img_url)
    
```
