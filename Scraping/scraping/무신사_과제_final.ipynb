```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_driver_user_agent():
    chrome_options = webdriver.ChromeOptions()
    
    chrome_options.add_argument("userAgent")
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=chrome_options
    )
    
    return driver
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


driver = get_driver_user_agent()
driver.get('https://www.musinsa.com/categories/item/002008')

# Wait for the page to load and display the items
wait = WebDriverWait(driver, 10)
item_element = wait.until(EC.visibility_of_element_located((By.ID, "searchList")))
li_elements = item_element.find_elements(By.CSS_SELECTOR, "li.li_box")
title_data = []
image_list=[]

# Extract item names from each element
for item_index in range(len(li_elements)):
    item = li_elements[item_index]
    try:
        attribute = item.find_element(By.CSS_SELECTOR, "div.article_info")
        item_brand = attribute.find_element(By.CSS_SELECTOR, "p.item_title").text
        item_title = attribute.find_element(By.CSS_SELECTOR, "p.list_info").text
        price = attribute.find_element(By.CSS_SELECTOR, "p.price").text.split()[-1]
        try:
            star = attribute.find_element(By.CSS_SELECTOR, "p.point > span.count").text
        except:
            star = None
        
        likes = attribute.find_element(By.CSS_SELECTOR, "p.txt_cnt_like").text
        likes = likes if likes else 0
        link = attribute.find_element(By.NAME, "goods_link").get_attribute("href")
        #상품 번호를 뽑아서 다시 
        driver.get(link)
        
        # 0인 값도 있으니, 고쳐야 한다
        season = driver.find_elements(By.CSS_SELECTOR, "li > p.product_article_contents")[1].text.split(' / ')[0]
        gender = driver.find_element(By.CSS_SELECTOR, "span.txt_gender").text
        image= driver.find_element(By.CSS_SELECTOR, "#detail_thumb ul.product_thumb li img").get_attribute('src')
        image_list.append(image)

        driver.back()
        item_element = wait.until(EC.visibility_of_element_located((By.ID, "searchList")))
        # 원래페이지로 돌아가고 나서 searchList를 찾기 전에 10초간 기다릴 수 있도록 하는 코드
        # 이걸 쓰지 않으면 driver.back()해서 원래 페이지에 돌아가도 아무것도 뜨지 않은 상태에서 크롤링 하려 하기 때문에 코드가 돌지 않음
        li_elements = item_element.find_elements(By.CSS_SELECTOR, "li.li_box")
        
        title_data.append([item_brand, item_title, price, star, likes,gender,season, image_list])
    except:
        continue

driver.quit()
print(title_data)
```

```python
import pandas as pd
df = pd.DataFrame(title_data)
df.columns=['브랜드','상품명','가격','좋아요','별','성별','시즌','이미지']
df
```
