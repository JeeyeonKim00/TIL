# **코스피, 코스닥, 달러환율 가져오고 데이터 프레임으로 만들기**
(2023~오늘까지)

- 코스피 가져오기

```python
import requests
KOSPI_URL="https://api.stock.naver.com/chart/domestic/index/KOSPI?periodType=dayCandle"
response = requests.get(KOSPI_URL)
response.status_code
```

```python
type(response)
```

```python
data = response.json()
data
```

```python
price_infos = data['priceInfos']
```

```python
import pandas as pd
df_kospi = pd.DataFrame(price_infos)
df_kospi  = df_kospi [df_kospi ['localDate']>='20230101']
df_kospi['localDate'] = pd.to_datetime(df_kospi['localDate'])
```

```python
df_kospi
```

- 코스닥 가져오기

```python
KOSDAQ_URL = "https://api.stock.naver.com/chart/domestic/index/KOSDAQ?periodType=dayCandle"
response = requests.get(KOSDAQ_URL)
response.text
```

```python
data =  response.json()
data
```

```python
price_info = data['priceInfos']
price_info
```

```python
df_kosdaq = pd.DataFrame(price_info)
df_kosdaq = df_kosdaq[df_kosdaq['localDate']>='20230101']
df_kosdaq['localDate'] = pd.to_datetime(df_kosdaq['localDate'])
df_kosdaq.info(0)
```

- 달러환율 가져오기

```python
USD_URL ="https://m.stock.naver.com/front-api/v1/chart/pricesByPeriod?category=exchange&reutersCode=FX_USDKRW&chartInfoType=marketindex&scriptChartType=areaYear"
response = requests.get(USD_URL)
response.status_code
```

```python
data = response.json()
data
```

```python
price_info = data['result']['priceInfos']
```

```python
df_usd = pd.DataFrame(price_info)

df_usd  = df_usd [df_usd ['localDate']>='20230101']
df_usd['localDate'] = pd.to_datetime(df_usd['localDate'])
df_usd
```

# **그래프 그리기**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import minmax_scale

plt.figure(figsize = (20,8))
plt.plot(df_kospi['localDate'], minmax_scale(df_kospi['closePrice']),'r--',label='kospi')
plt.plot(df_kosdaq['localDate'], minmax_scale(df_kosdaq['closePrice']),'g:',label='kosdaq')
plt.plot(df_usd['localDate'], minmax_scale(df_usd['closePrice']),'b^--', label='usd');
plt.xticks(df_usd['localDate'],rotation=45, fontsize=7)
plt.legend()
plt.show()
```
