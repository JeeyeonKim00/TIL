```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
iris = load_iris()

irisDF = pd.DataFrame(iris.data, columns= iris.feature_names)
irisDF['target'] = iris.target
```

```python
irisDF
```

# GMM

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components = 3, random_state =0).fit(iris.data)
gmm_cluster_labels = gmm.predict(iris.data)
```

```python
gmm = GaussianMixture(n_components = 3, random_state =0)
gmm_cluster_labels = gmm.fit_predict(iris.data)
gmm_cluster_labels
```

```python
irisDF['gmm_cluster'] = gmm_cluster_labels 
```

```python
irisDF['target'] = iris.target
```

```python
irisDF
```

```python
iris_result = irisDF.groupby(['target'])['gmm_cluster'].value_counts()
iris_result
```

```python
irisDF.groupby(['target','gmm_cluster']).count()['sepal length (cm)']
```

# KMeans

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters =3, init='k-means++',max_iter = 300, random_state=0)
kmeans_cluster_labels = kmeans.fit_predict(iris.data)
kmeans_cluster_labels
```

```python
irisDF2 = pd.DataFrame(iris.data, columns = iris.feature_names)
irisDF2['kmeans_cluster'] = kmeans_cluster_labels
irisDF2['target'] = iris.target
irisDF2
```

```python
irisDF2.groupby(['target','kmeans_cluster']).count()['sepal length (cm)']
```
