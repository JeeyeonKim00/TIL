# 데이터 전처리
- 선형모델을 사용할 때는 표준화를 진행하는 것이 좋다.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
data_scaled = scaler.fit_transform(cancer.data)

X_train, X_test, y_train, y_test = train_test_split(
        data_scaled,
        cancer.target,
        test_size = 0.3,
        random_state =0)
```

```python
lr_clf = LogisticRegression() # 기본 solver = lbfgs
lr_clf.fit(X_train, y_train)

lr_pred = lr_clf.predict(X_test)
```

# 평가

```python
from sklearn.metrics import accuracy_score, roc_auc_score

print('Accuracy: {:.3f}, ROC_AUC: {:.3f}'.format(accuracy_score(lr_pred,y_test),roc_auc_score(lr_pred, y_test)))
```

## solver에 따른 성능 변화 측정

```python
solvers = ['lbfgs','liblinear','newton-cg','sag','saga']

for solver in solvers:
    lr_clf=LogisticRegression(solver=solver, max_iter=600) #max_iter = 최적화 횟수(최대 600번이라는 의미)
    lr_clf.fit(X_train, y_train)
    
    lr_pred = lr_clf.predict(X_test)
    
    print(solver)
    print('Accuracy: {:.3f}, ROC_AUC: {:.3f}'.format(accuracy_score(lr_pred,y_test),roc_auc_score(lr_pred, y_test)))
    print()
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV

params={'solver':['liblinear', 'lbfgs'],
        'penalty':['l2', 'l1'],
        'C':[0.01, 0.1, 1, 1, 5, 10]}

grid_clf = GridSearchCV(
        LogisticRegression(),
        param_grid = params,
        scoring='accuracy',
        cv=3)

grid_clf.fit(data_scaled, cancer.target)
```

```python
grid_clf.best_params_
```

```python
print('''최적 하이퍼 파라미터:{}, 
최적 평균 정확도:{:.3f}'''.format(grid_clf.best_params_, 
                                                  grid_clf.best_score_))
```
