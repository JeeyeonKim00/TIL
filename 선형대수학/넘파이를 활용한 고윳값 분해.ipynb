# np.linalg.eig -> 고윳값 분해 함수

```python
A = np.array([[1,-2],
             [2,-3]])

l1, V1 = np.linalg.eig(A)

print('lambda : {}'.format(l1),end='\n\n')
print('V : \n{}'.format(V1))
```

```python
B = np.array([[2,3],
              [2,1]])

l2, V2 = np.linalg.eig(B)

print('lambda  : {}'.format(l2),end='\n\n')
print('V : \n{}'.format(V2))
```

- 복소수 고유값, 복소수 고유벡터

```python
C = np.array([[0,-1],
             [1,0]])

l3, V3 = np.linalg.eig(C)

print('lambda  : {}'.format(l3),end='\n\n')
print('V : \n{}'.format(V3))
```

# 대각화

```python
B = np.array([[2,3],
              [2,1]])

lambda_b, V = np.linalg.eig(B)

lambda_b, V

# lambda_b : 고윳값 벡터
# V: 고윳값 행렬
```

```python
# 고윳값 행렬 만들기
L = np.diag(lambda_b)
L
```

```python
# 고유벡터 행렬의 역행렬
V_inv = np.linalg.inv(V)
V_inv
```

```python
V @ L @ V_inv # = B
```

# 대칭행렬을 랭크-1 행렬의 합으로 분해

```python
A = np.array([[60., 30., 20.], [30., 20., 15.], [20., 15., 12.]])
A
```

```python
w, V = np.linalg.eig(A)
```

```python
w # 고윳값 벡터
```

```python
V # 고유벡터행렬
```

```python
# 각각의 고유값 구하기
w1, w2, w3= w

# 각각의 고유벡터 구하기
v1 = V[:,0:1]
v2 = V[:,1:2]
v3 = V[:,2:3]

v1.shape, v2.shape, v3.shape
```

```python
A1 = v1 @ v1.T
A2 = v2 @ v2.T
A3 = v3 @ v3.T
```

```python
w1* A1
```

```python
w2 * A2
```

```python
w3 * A3
```

```python
w1*A1 + w2*A2
```

```python
w1*A1 + w2*A2 + w3*A3
```

```python
# A의  역행렬
np.linalg.inv(A)
```

```python
# 고윳값의 역수를 구해서 다 더해주면 된다.
1/w1 * A1 + 1/w2 * A2 + 1/w3 * A3
```
