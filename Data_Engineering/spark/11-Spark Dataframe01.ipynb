- SQLì˜ count case when ê³¼ ìœ ì‚¬í•œ ë¡œì§ìœ¼ë¡œ null ê°’ì„ í¬í•¨í•œ ì»¬ëŸ¼ í™•ì¸
- What's the problem? íŒŒì´ì¬ì—ì„œ Nullê°’ì„ NaN, Nullì„ ëª¨ë‘ ì‚¬ìš©í•œë‹¤ëŠ” ê²ƒ


```python
from pyspark.sql.functions import count, isnan, when, col
```

```python
# Spark DFì—ì„œ ì»¬ëŸ¼ë§Œ í™•ì¸
titanic_sdf.columns
```

```python
# ìœ„ì˜ ë‚˜ì˜¨ ì»¬ëŸ¼ ë³„ë¡œ NaNê°’ì˜ ì¡´ì¬ ì—¬ë¶€ íŒŒì•… ë° ê°œìˆ˜ë¥¼ ì„¸ì•¼ í•œë‹¤.
titanic_sdf.select([count(when(isnan(c)|col(c).isNull(),c)).alias(c) for c in titanic_sdf.columns]).show()
```

**comprehension tip**
- comprehension ì½ì„ ë•ŒëŠ” ë¬´ì¡°ê±´ `in` ë’¤ë¶€í„° ì½ëŠ”ë‹¤.
- 

```python
def check_spark_null(spark_df):
    return spark_df.select([count(when(isnan(c)|col(c).isNull(),c)).alias(c) for c in titanic_sdf.columns])

check_spark_null(titanic_sdf).show()
```

# spark DataFrame filter() ë©”ì†Œë“œ ì•Œì•„ë³´ê¸°
- filter()ëŠ” SQLì˜ `where`ì™€ ìœ ì‚¬í•˜ê²Œ DataFrame ë‚´ì˜ íŠ¹ì • ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ë ˆì½”ë“œë¥¼ DataFrameìœ¼ë¡œ ë°˜í™˜.

- filter() = where()
- where() ë©”ì†Œë“œëŠ” filter()ì˜ aliasì´ë©° SQL whereì™€ ì§ê´€ì ì¸ ë™ì¼ì„±ì„ ê°–ì¶”ê¸° ìœ„í•´ ìƒì„±.
- ë³µí•© ì¡°ê±´ andëŠ” `&` ë¡œ, orë¥¼ `|`ë¡œ ì‚¬ìš©. ê°œë³„ì¡°ê±´ì€ ()ë¡œ ê°ì‹¸ì•¼ í•¨.
