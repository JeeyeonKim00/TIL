```python
import pyspark
pyspark.__version__
```

```python
from pyspark import SparkConf, SparkContext
```

```python
# 스파크 환경 설정 객체 생성
conf = SparkConf().setMaster('local').setAppName('country-student-counts') # Hadoop쓰지 않고 local 컴에서 해결

# SparkContext 생성
# 변수명은 항상 sc로 할 것
sc = SparkContext(conf = conf)
```

```python
!pwd
```

```python
filepath = '/home/ubuntu/working/spark-examples/data/xAPI-Edu-Data.csv'
```

```python
# 로컬 파일에서 데이터를 불러올 것이기 때문에 shceme이 file://
# 만약 HDFS에서 불러온다? hdfs://...
# 로컬에서 불러온 textfile이 RDD가 됨.

lines = sc.textFile(f'file:///{filepath}') 
lines
```

```python
header = lines.first()
header
```

```python
# RDD에는 컬럼명이 필요없으니까 헤더 제외할 것
# 헤더 제외한 데이터만 불러오기
# filter는 transformation. - 데이터 변경하는거니까

datas = lines.filter(lambda row : row != header)  # filter : Transformation
datas
```

```python
# 갖고있는 데이터를 펼쳐놓는 과정 = map

countries = datas.map(lambda row : row.split(',')[2]) # map: Transformation
countries
```

Transformation의 결과물 = RDD
Action의 결과물 = 값

```python
countries.collect() # Action. 모든 Transformation 작업을 한꺼번에 수행
```

```python
result = countries.countByValue() # 나라별로 몇개의 데이터가 있나?
result
```

```python
# 작업이 끝났으면 반드시 스파크 컨텍스트 종료
sc.stop()
```
