# 코스피 정보

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd

KOSPI_URL = "https://api.stock.naver.com/chart/domestic/index/KOSPI?periodType=dayCandle"
response = requests.get(KOSPI_URL)
data = response.json()
price_info=data['priceInfos']
df_kospi = pd.DataFrame(price_info)
df_kospi
```

# 코스닥 정보 

```python
KOSDAQ_URL ="https://api.stock.naver.com/chart/domestic/index/KOSDAQ?periodType=dayCandle"
response_kosdaq = requests.get(KOSDAQ_URL)
data_k = response_kosdaq.json()
price_info_kosdaq=data_k['priceInfos']
df_kosdaq = pd.DataFrame(price_info_kosdaq)
df_kosdaq
```

# 환율

```python
USD_URL="https://m.stock.naver.com/front-api/v1/chart/pricesByPeriod?category=exchange&reutersCode=FX_USDKRW&chartInfoType=marketindex&scriptChartType=areaYear"
response_usd = requests.get(USD_URL)
data_u = response_usd.json()
price_info_usd=data_u['result']['priceInfos']
df_usd = pd.DataFrame(price_info_usd)
df_usd
```

# 데이터 편집

```python
df_kospi = df_kospi[df_kospi['localDate']>='20230101']
df_kosdaq = df_kosdaq[df_kosdaq['localDate']>='20230101']
df_usd = df_usd[df_usd['localDate']>='20230101']

df_kospi
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
%matplotlib inline
```

```python
plt.figure(figsize=(20,8))
fig = plt.plot(df_kospi['localDate'],minmax_scale(df_kospi['closePrice']),label='kospi')
fig = plt.plot(df_kosdaq['localDate'],minmax_scale(df_kosdaq['closePrice']),label='kosdaq')
fig = plt.plot(df_usd['localDate'],minmax_scale(df_usd['closePrice']),label='usd')


plt.xticks(df_kospi['localDate'],rotation=45,fontsize=7)
plt.legend()
```
