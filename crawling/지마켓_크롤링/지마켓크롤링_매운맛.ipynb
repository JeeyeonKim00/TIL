# 전체 URL 뽑기

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

GMARKET_BEST_PRODUCT_URL = "https://corners.gmarket.co.kr/BestSellers"
res = requests.get(GMARKET_BEST_PRODUCT_URL)

soup=BeautifulSoup(res.content,'html.parser')


url_category_list=[]
for i in range(13):
    url= "https://corners.gmarket.co.kr/"+soup.select('#categoryTabG > li>a')[i]['href']
    category = soup.select("#gBestWrap #categoryTabG > li>a")[i].text
    url_category_list.append([url,category])
print(url_category_list)
```

```python
for url,category in url_category_list:
    print(url,category)
```

```python
product_code_list=[]
img_url_list=[]
dc_price_list=[]
detail_page_list=[]
product_name_list=[]
rank_list=[]
category_list=[]



for URL, CATEGORY in url_category_list:
    for i in range(0,100):
        res = requests.get(URL)
        soup = BeautifulSoup(res.content, 'html.parser')
        best_item= soup.select("#gBestWrap > div > div:nth-child(5) > div > ul >li")

        category = CATEGORY
        rank= best_item[i].select_one("p").text
        product_name = best_item[i].find("a",{'class':'itemname'}).text
        detail_page = best_item[i].select_one('a')['href']
        product_code = best_item[i].select_one('a')['href'].split('=')[1][0:10]
        img_url = "https:"+best_item[i].select_one('img.lazy')['data-original']
        dc_price = best_item[i].select_one("div.s-price > strong > span").text

        if not(dc_price):
            dc_price = best_item[i].select_one('div.o-price > span:nth-child(2)').text


        product_code_list.append(product_code)
        product_name_list.append(product_name)
        detail_page_list.append(detail_page)
        img_url_list.append(img_url)
        dc_price_list.append(dc_price)
        rank_list.append(rank)
        category_list.append(category)

gmarket_datas ={'순위': rank_list,'카테고리':category_list,'이미지url':img_url_list,
               '할인가':dc_price_list,'상세 페이지':detail_page_list,
               '상품명':product_name_list,'상품코드':product_code_list}
```

# 뽑은 gmarket_datas 를 데이터프레임으로 전환하기

```python
import pandas as pd
df = pd.DataFrame(gmarket_datas)
df['할인가']=df['할인가'].str.rstrip('원').str.replace(',','')
df
```

```python
df['할인가'] = df['할인가'].astype(int)
```

```python
df['순위'] = df['순위'].astype(int)
```

```python
df.info()
```

# df.groupby().mean()으로 평균

```python
df_top50 = df[df['순위']<=50]
df_bottom50 = df[df['순위']>50]

df_top50_mean = df_top50.groupby('카테고리').mean()
df_bottom50_mean = df_bottom50.groupby('카테고리').mean()
```
