```python
#자동차 관리 프로그램 - 변수 작업

종료조건= True

while 종료조건: #화면 출력
    print()
    print("#### 자동차 정보 프로그램 ####")
    print("1.자동차 정보 입력")
    print("2.저장된 목록 보기")
    print("3.각 자동차의 구매가격 조회")
    print("4.옵션이 있는 자동차 정보 조회")
    print("5.빠른 자동차와 느린 자동차의 속도차 조회")
    print("6.프로그램 이용 종료")
    번호=input('입력:')[0] #문자
    
    
    #동작기능
    if 번호 =='1':
        이름=input('이름: ')
        엔진가격 = int(input('엔진가격: '))
        타이어가격 = int(input('타이어 가격:'))
        옵션 = input('옵션(있다/없다): ')
        옵션이름 = input('옵션 이름: ')
        최고속도 = input('최고 속도')
        구매가격 = 엔진가격 + 타이어가격
        print('차량정보 입력 완료')
        
    elif 번호 =='2':
        print(이름, 엔진가격, 타이어가격, 옵션,옵션이름,최고속도,구매가격)
        print('저장된 목록 출력 완료')
    elif 번호 =='3':
        print(이름,구매가격)
        print('각 자동차 구매가격 출력 완료')
    elif 번호 =='4':
        if 옵션 =='있다':
            print(이름, 엔진가격, 타이어가격, 옵션,옵션이름,최고속도,구매가격)
        print('저장된 목록 출력 완료')
            
    elif 번호 =='5':
        print('가장 빠른차 속도 - 가장 느린차 속도')
        print('속도차 출력 완료')
    elif 번호 =='6':
        종료조건=False
        print()
    else:
        print("입력은 1~6번까지만 기능이 있습니다. \n다시 입력해주세요.")
else:
    print('프로그램이 정상적으로 종료됩니다.')
```

```python
# 자동차 관리 프로그램 - 1차 리스트 작업 (모델1)

종료조건= True

 #화면 출력
화면출력 = '''#### 자동차 정보 프로그램 ####
    1.자동차 정보 입력
    2.저장된 목록 보기
    3.각 자동차의 구매가격 조회
    4.옵션이 있는 자동차 정보 조회
    5.빠른 자동차와 느린 자동차의 속도차 조회
    6.프로그램 이용 종료
    입력:'''

자동차_리스트=[]
자동차번호=0
최고속도=0
최저속도=501
#여기서 입력가능한 속도 범위는 0~500
while 종료조건:
    번호=input('입력:')[0] #문자
    
    
    #동작기능
    if  번호 =='1':
        자동차_리스트.append(input('이름')) #append는 뒤에다가 붙이기 시작 #인덱스0
        자동차_리스트.append(int(input('엔진가격: ')))#인1
        자동차_리스트.append(int(input('타이어 가격:')))#인2
        자동차_리스트.append(input('옵션(있다/없다): '))#인3
        if 자동차_리스트[(자동차번호*7)+3]=='있다':
            자동차_리스트.append(input('옵션 이름: '))#인4
        else:
            자동차_리스트.append('없음')
        자동차_리스트.append(input('최고 속도'))#인5
        최고속도 = 최고속도 if 자동차_리스트[(자동차번호*7)+5] <최고속도 else 자동차_리스트[(자동차번호*7)+5] 
        최저속도 = 최저속도 if 자동차_리스트[(자동차번호*7)+5] >최저속도 else 자동차_리스트[(자동차번호*7)+5]
        자동차_리스트.append(자동차_리스트[(자동차번호*7)+1]+자동차_리스트[(자동차번호*7)+2])
        자동차번호+=1
        print('차량정보 입력 완료')
        
    elif 번호 =='2':
        for i in range(자동차번호):
            print(f'{i+1}번째 차량')
            print(f'''이름:{자동차_리스트[i*7]},
엔진가격:{자동차_리스트[i*7]+1},
타이어가격 :{자동차_리스트[i*7]+2}, 
옵션: {자동차_리스트[i*7]+3},
옵션이름: {자동차_리스트[i*7]+4},
최고속도: {자동차_리스트[i*7]+5},
구매가격: {자동차_리스트[i*7]+6}''')
        print('저장된 목록 출력 완료')
    elif 번호 =='3':
        for i in range(자동차번호):
            print(f'{i+1}번째 차량')
            print(f'''이름:{자동차_리스트[i*7]},구매가격: {자동차_리스트[i*7]+6}''')
        print('각 자동차 구매가격 출력 완료')
    elif 번호 =='4':
        for i in range(자동차번호):
                if 옵션 =='있다':
                    print(f'''이름:{자동차_리스트[i*7]},
엔진가격:{자동차_리스트[i*7]+1},
타이어가격 :{자동차_리스트[i*7]+2}, 
옵션: {자동차_리스트[i*7]+3},
옵션이름: {자동차_리스트[i*7]+4},
최고속도: {자동차_리스트[i*7]+5},
구매가격: {자동차_리스트[i*7]+6}''')
        print('저장된 목록 출력 완료')
            
    elif 번호 =='5':
        print(f'가장 빠른차 속도 - 가장 느린차 속도={최고속도-최저속도}')
        print('속도차 출력 완료')
    elif 번호 =='6':
        종료조건=False
        print()
    else:
        print("입력은 1~6번까지만 기능이 있습니다. \n다시 입력해주세요.")
else:
    print('프로그램이 정상적으로 종료됩니다.')
```
