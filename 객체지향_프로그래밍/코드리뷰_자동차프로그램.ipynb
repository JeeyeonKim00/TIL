```python
종료조건= True
화면출력='''###자동차 정보 프로그램###
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료'''

while 종료조건:
    print(화면출력)
    번호 =input('입력: ')[0]
    
    if 번호 =='1':
        이름 = input('이름: ')
        엔진가격 = int(input('엔진가격: '))
        타이어가격 = int(input('타이어가격: '))
        옵션 = input('옵션(있다/없다): ')
        옵션이름 = input('옵션이름: ')
        최고속도 = int(input('최고속도: '))
        구매가격= 엔진가격 +타이어가격
        print('차량정보 입력 완료')
        
    elif 번호 =='2':
        print(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도,구매가격)
        print('저장된 목록 출력 완료')
    elif 번호 =='3':
        print(이름,구매가격)
        print('각 자동차 구매가격 출력 완료')
    elif 번호 =='4':
        if 옵션=='있다':
            print(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도,구매가격)
            print('저장된 목록 출력완료')
    elif 번호 =='5':
        print('가장 빠른차속도-가장느린차 속도 ')
        print('속도차 출력 완료')
    elif 번호 =='6':
        print('프로그램이 종료 됩니다.')
        종료조건 =False
    else:
        print("입력은 1~6번까지만 가능합니다.\n다시 입력해주세요.")
else:
    print('프로그램이 정상적으로 종료됩니다.')
    
```

# 3번 리스트2

```python
종료조건= True
화면출력='''###자동차 정보 프로그램###
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료'''

자동차_리스트=[]
자동차번호=0
최고속도=0
최저속도=501

while 종료조건:
    print(화면출력)
    번호 =input('입력: ')[0]
    
    if 번호 =='1':
        자동차_정보=[]
        자동차_정보.append(input('이름: ')) #인덱스0
        자동차_정보.append(int(input('엔진가격: '))) #인덱스1
        자동차_정보.append(int(input('타이어가격: '))) #인덱스2
        자동차_정보.append(input('옵션(있다/없다): ')) #3
        if 자동차_정보[3]=='있다':
            자동차_정보.append(input('옵션이름: ')) #4
        else:
            자동차_정보.append('없음')
        자동차_정보.append(int(input('최고속도: ')))#5
        최고속도=최고속도 if 최고속도>자동차_정보[5] else 자동차_정보[5]
        최저속도=최저속도 if 최저속도<자동차_정보[5] else 자동차_정보[5]
        
        자동차_정보.append(자동차_정보[1]+자동차_정보[2]) #6
        자동차_리스트.append(자동차_정보)
        print('차량정보 입력 완료')
        print('----------------------------------')

    elif 번호 =='2':
        기록번호=0
        for 자동차 in 자동차_리스트:
            기록번호+=1
            print(f'{기록번호}번째 차량')
            print(f'''이름: {자동차[0]}
엔진가격: {자동차[1]}
타이어가격: {자동차[2]}
옵션: {자동차[3]}
옵션이름: {자동차[4]}
최고속도: {자동차[5]}
구매가격: {자동차[6]}''')
        print('저장된 목록 출력 완료')
        print('----------------------------------')
        
    elif 번호 =='3':
        기록번호=0
        for 자동차 in 자동차_리스트:
            기록번호+=1
            print(f'{기록번호}번째 차량')
            print(f'''이름: {자동차[0]}
구매가격: {자동차[6]}''')
        print('각 자동차 구매가격 출력 완료')
        print('----------------------------------')
        
    elif 번호 =='4':#옵션이 있는 자동차 정보를 조회
        for 기록번호, 자동차 in [t for t in [(i+1,v) for i,v in enumerate(자동차_리스트)] if t[1][3]=='있다']:
            print(f'{기록번호}번째 차량')
            print(f'''이름: {자동차[0]}
엔진가격: {자동차[1]}
타이어가격: {자동차[2]}
옵션: {자동차[3]}
옵션이름: {자동차[4]}
최고속도: {자동차[5]}
구매가격: {자동차[6]}''')
        print('저장된 목록 출력완료')
        print('----------------------------------')

    elif 번호 =='5':
        print(f'{최고속도-최저속도}')
        print('속도차 출력 완료')
        print('----------------------------------')

    elif 번호 =='6':
        print('프로그램이 종료 됩니다.')
        print('----------------------------------')
        종료조건 =False
    
    else:
        print("입력은 1~6번까지만 가능합니다.\n다시 입력해주세요.")

else:
    print('프로그램이 정상적으로 종료됩니다.')
    
```

# 4번: 딕셔너리1

```python
종료조건= True
화면출력='''###자동차 정보 프로그램###
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료'''

자동차_리스트=[]
자동차번호=0
최고속도=0
최저속도=501

while 종료조건:
    print(화면출력)
    번호 =input('입력: ')[0]
    
    if 번호 =='1':
        자동차_정보={}
        자동차_정보['이름']=input('이름: ') #인덱스0
        자동차_정보['엔진가격']=int(input('엔진가격: ')) #인덱스1
        자동차_정보['타이어가격']=int(input('타이어가격: ')) #인덱스2
        자동차_정보['옵션']=input('옵션(있다/없다): ') #3
        if 자동차_정보['옵션']=='있다':
            자동차_정보['옵션이름']=input('옵션이름: ') #4
        else:
            자동차_정보['옵션이름']='없음'
        자동차_정보['최고속도']=int(input('최고속도: ')) #5
        최고속도=최고속도 if 최고속도>자동차_정보['최고속도'] else 자동차_정보['최고속도']
        최저속도=최저속도 if 최저속도<자동차_정보['최고속도'] else 자동차_정보['최고속도']
        
        자동차_정보['구매가격']=자동차_정보['엔진가격']+자동차_정보['타이어가격'] #6
        자동차_리스트.append(자동차_정보)
        print('차량정보 입력 완료')
        print('----------------------------------')

    elif 번호 =='2':
        기록번호=0
        for 자동차 in 자동차_리스트:
            기록번호+=1
            print(f'{기록번호}번째 차량')
            print(f'''이름: {자동차[0]}
엔진가격: {자동차[1]}
타이어가격: {자동차[2]}
옵션: {자동차[3]}
옵션이름: {자동차[4]}
최고속도: {자동차[5]}
구매가격: {자동차[6]}''')
        print('저장된 목록 출력 완료')
        print('----------------------------------')
        
    elif 번호 =='3':
        기록번호=0
        for 자동차 in 자동차_리스트:
            기록번호+=1
            print(f'{기록번호}번째 차량')
            print(f'''이름: {자동차[0]}
구매가격: {자동차[6]}''')
        print('각 자동차 구매가격 출력 완료')
        print('----------------------------------')
        
    elif 번호 =='4':#옵션이 있는 자동차 정보를 조회
        for 기록번호, 자동차 in [t for t in [(i+1,v) for i,v in enumerate(자동차_리스트)] if t[1][3]=='있다']:
            print(f'{기록번호}번째 차량')
            print(f'''이름: {자동차[0]}
엔진가격: {자동차[1]}
타이어가격: {자동차[2]}
옵션: {자동차[3]}
옵션이름: {자동차[4]}
최고속도: {자동차[5]}
구매가격: {자동차[6]}''')
        print('저장된 목록 출력완료')
        print('----------------------------------')

    elif 번호 =='5':
        print(f'{최고속도-최저속도}')
        print('속도차 출력 완료')
        print('----------------------------------')

    elif 번호 =='6':
        print('프로그램이 종료 됩니다.')
        print('----------------------------------')
        종료조건 =False
    
    else:
        print("입력은 1~6번까지만 가능합니다.\n다시 입력해주세요.")

else:
    print('프로그램이 정상적으로 종료됩니다.')
    
```

```python
#1.입력 /def 입력()
#2.목록보기 /def 목록출력()
#3.구매가격 조회 /def 가격출력()
#4.자동차 옵션 정보 조회 /def 옵션_정보_출력)_
#5.빠른자동차, 느린 자동차의 속도차 조회  /def 속도차_출력
#6. 프로그램 이용 종료/ def 종료

def 입력():
    global 리스트_최고속도
    global 리스트_최저속도
    이름=input('이름: ')
    엔진가격= int(input('엔진가격: '))
    타이어가격=int(input('타이어가격: '))
    옵션=input('옵션: ')
    if 옵션=='있다':
        옵션이름=input('옵션이름: ')
    else:
        옵션이름='없다'
    최고속도=int(input('최고속도: '))
    리스트_최고속도=리스트_최고속도 if 리스트_최고속도>최고속도 else 최고속도
    리스트_최저속도=리스트_최저속도 if 리스트_최저속도<최고속도 else 최고속도
    
    자동차_정보=자동차정보(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도)
    자동차_리스트.append(자동차_정보)
    print('차량 정보입력 완료')
    print('-'*30)
    
def 목록출력():
    for 입력번호, 자동차 in [(i+1,v) for i,v in enumerate(자동차_리스트)]:
        print(f'{입력번호}번|{자동차}')
        print('-'*30)
    print('저장된 목록 출력 완료')
    
def 가격출력():
    for 입력번호,자동차 in [(i+1,v) for i,v in enumerate(자동차_리스트)]:
        print(f'{입력번호}번|{자동차.print_구매가격()}')
        print('-'*30)
    print('각 자동차 구매가격 출력 완료')

def 옵션_정보_출력():
    for 입력번호, 자동차 in [t for t in [(i+1,v) for i,v in enumerate(자동차_리스트)] if t[1].옵션=='있다']:
        print(f'{입력번호}번|{자동차}')
        print('-'*30)
        print('저장된 목록 출력 완료')
        
def 속도차_출력():
    print(f'{리스트_최고속도-리스트_최저속도}')
    print('속도차 출력 완료')
    
def 종료():
    print('프로그램이 종료됩니다,')
    return True

def UI():
    화면출력='''###자동차 정보 프로그램###
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료'''
    return input(화면출력)[0]    

    
```

# classmethod사용

```python
class 자동차정보:
    자동차_리스트=[]
    리스트_최고속도=0
    리스트_최저속도=501
    
    def __init__(self,이름,엔진가격,타이어가격,옵션,옵션이름,최고속도):
        self.이름 = 이름
        self.엔진가격 = 엔진가격
        self.타이어가격= 타이어가격
        self.옵션= 옵션
        self.옵션이름 = 옵션이름
        self.최고속도 = 최고속도
        self.구매가격 =엔진가격+타이어가격
    def __str__(self):
        return f'이름:{self.이름}|엔진가격:{self.엔진가격}|타이어가격:{self.타이어가격}|옵션:{self.옵션}|옵션이름:{self.옵션이름}|최고속도:{self.최고속도}'
    def print_구매가격(self):
        return f'이름:{self.이름}|구매가격:{self.구매가격}'
    
    @classmethod
    def 입력(cls):
        이름 = input('이름: ')
        엔진가격=int(input('엔진가격: '))
        타이어가격=int(input('타이어가격: '))
        옵션 = input('옵션(있다/없다): ')
        if 옵션=='있다':
            옵션이름=input('옵션이름: ')
        else:
            옵션이름='없다'
        최고속도= int(input('최고속도: '))
        cls.리스트_최고속도= cls.리스트_최고속도 if 최고속도 <cls.리스트_최고속도 else 최고속도
        cls.리스트_최저속도= cls.리스트_최저속도 if 최고속도 >cls.리스트_최저속도 else 최고속도
        
        자동차_정보=자동차정보(이름,엔진가격,타이어가격,옵션,옵션이름,최고속도)
        cls.자동차_리스트.append(자동차_정보)
        print('-'*30)
        print('차량정보 입력 완료')
    
    @classmethod
    def 목록출력(cls):
        for 입력번호,자동차 in [(i+1,v) for i,v in enumerate(cls.자동차_리스트)]:
            print(f'{입력번호}번|{자동차}')
            print('-'*30)
        else:
            print('저장된 목록 출력 완료')
            
    @classmethod
    def 가격출력(cls):
        for 입력번호,자동차 in [(i+1,v) for i,v in enumerate(cls.자동차_리스트)]:
            print(f'{입력번호}번|{자동차.print_구매가격()}')
            print('-'*30)
        else:
            print('각 자동차 구매가격 출력 완료')
            
    @classmethod
    def 옵션_정보_출력():
        for 입력번호, 자동차 in [t for t in [(i+1,v) for i,v in enumerate(cls.자동차_리스트)] if t[1].옵션=='있다']:
            print(f'{입력번호}번|{자동차}')
            print('-'*30)
        else:
            print('저장된 목록 출력 완료')
            
    @classmethod
    def 속도차_출력(cls):
        print(f'{cls.리스트_최고속도-cls.리스트_최저속도}')
        print('속도차 출력 완료')
        
    @classmethod
    def 종료(cls):
        print('프로그램이 종료됩니다.')
        return True
    
    @classmethod
    def UI(cls):
        화면출력='''###자동차 정보 프로그램###
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 프로그램 이용 종료'''
        return input(화면출력)[0]
    
run=False
while not run:
    n=자동차정보.UI()
    if n not in '123456':
        print(동작리스트)
    else:
        run=동작리스트[int(n)]()
    
        
        
```
